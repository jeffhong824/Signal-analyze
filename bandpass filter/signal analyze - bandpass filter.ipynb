## 通用 bandpass filter 
### 推薦 非edf訊號 使用 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
path = 'C:/Users/student/Desktop/signal-analysis/'
```

```python
filter_before = np.load( path + 'epilepsy_after_filter.npy')
```

```python
filter_before.shape
```

```python
def bandpass_filter(data,lowcut,highcut,sampling_freq,duration): #lowcut highcut sampling_freq sec
    
    t = np.arange(0.0, duration, 1/sampling_freq) 
    filter = signal.firwin(numtaps=400, cutoff=[lowcut/(sampling_freq/2), highcut/(sampling_freq/2)], pass_zero=False) 

    filter_after = signal.convolve(data, filter, mode='same')
    
    return filter_after

filter_after = bandpass_filter(filter_before,1,50,250,8)
```

```python
plt.plot(filter_before)
plt.show()
```

```python
plt.plot(filter_after)
plt.show()
```
